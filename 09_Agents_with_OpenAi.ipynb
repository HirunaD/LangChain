```python
from transformers import AutoModelForCausalLM, pipeline
from langchain_community.llms import HuggingFacePipeline
import torch
import os

# Set Hugging Face API key
os.environ["HUGGINGFACEHUB_API_TOKEN"] = userdata.get('HUGGINGFACE_API_KEY')

# model_id = "tiiuae/falcon-7b-instruct"
model_id = "mistralai/Mistral-7B-v0.1"

model = AutoModelForCausalLM.from_pretrained(model_id, device_map="auto", torch_dtype=torch.bfloat16)

# Initialize the model from Hugging Face Hub
pipe = pipeline(
    model_id="mistralai/Mistral-7B-v0.1",  # Replace with your desired model
    model_kwargs={"temperature": 0.0, "max_new_tokens": 256}
)

llm = HuggingFacePipeline(pipeline=pipe)
```
